```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Exploración inicial del dataset (dimensiones, tipos y nulos)
print(df.shape)          
print(df.dtypes)         
print(df.isnull().sum()) 

# Análisis del balanceo de clases en la variable objetivo ('Churn')
df["Churn"].value_counts(normalize=True).plot(kind="bar")
plt.title("Distribución de Churn")
plt.show()

# Análisis de correlación lineal entre variables numéricas
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.show()

# Distribución de frecuencia de las variables numéricas
df.select_dtypes(include="number").hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```
